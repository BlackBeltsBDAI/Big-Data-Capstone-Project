# Kafka : Consume data from topic

Imports

```python
from pykafka import KafkaClient
import json
from pandas import json_normalize
```

Function to initialize kafka client

```python
def get_kafka_client():
    return KafkaClient(hosts='127.0.0.1:9092')
```

Initialize kafka client

```python
client = get_kafka_client()
```

Specify the topic that we want to subscribe to or receive data from, which is "Police.Crimes"

```python
topic = client.topics['Police.Crimes']
```

Create consumer which will be used to read crime records from the topic

```python
consumer = topic.get_simple_consumer()
```

Read and print messages coming from the topic

```python
for message in consumer:
    if message is not None:
        message_dict = json.loads('{' + str(message.value).split('{', 1)[1][:-1])
        message_df = json_normalize(message_dict['fullDocument'])
        print(message_df.squeeze())
```
